```python
'''
write to table header
'''

with open("../../dataset/merge/2021_dateset.csv", "a", encoding='utf-8', newline='') as fs:
        fieldnames = ['no','gender(female:0,male:1)','age','height(cm)','weight(Kg)','bmi','season',
                      'preference(-1:cool 0:n 1:warm)','sensitivity(0:in 1:slight 2:very)','environment assess(-1:c 0:n 1:w)','date','time','room',
                      'thermal sensation','thermal comfort (1:com,0:uncom)','thermal preference(-1:c,0:not change,1:w)','temp','humid']
        writer = csv.DictWriter(fs, fieldnames=fieldnames)
        writer.writeheader()
```

```python
'''
read excel
'''
summer = pd.read_csv('../../dataset/Handling data set/2021_summer.csv',encoding="gbk")
winter = pd.read_csv('../../dataset/Handling data set/2021_winter.csv',encoding="gbk")
names = ['边策','卞云山','曹天翔','柴臻豪','程延','董翔翔','杜帅飞','段婷芳','付业琛',
         '高泱晗','顾腾','顾妍','何向萌','贺思哲','侯江伟','吉睿琨','李佳欣','李新路',
         '李星奇','李杨','李宇婷','李子文','刘家丞','梁天堡','刘亚萍','吕晓亮','乔潇逸',
         '邵柯','苏莹','邵世彪','申沅均','沈婷','田颖','王青乙','王洁','王秦龙','王姝婕',
         '王钰琳','魏冬阳','魏佳利','魏亚鹏','武龙坤','杨晨阳','杨雨洁','余俊','余香南',
         '袁丹夫','张源鸿','张栋','张艺潇','张远东','张悦仙','赵浩铭','赵国梁','郑超',
         '周春翔','赵艳玲','朱济琛']

winter.dropna(axis=0, how='any',inplace = True)
summer.dropna(axis=0, how='any',inplace = True)
print(summer)
print(winter)

# 将summer 与 winter数据中的姓名字段替换为编号
for i in range(len(names)):
    summer.replace({'name':{names[i]:i}},inplace = True)
    winter.replace({'subject':{names[i]:i}},inplace = True)

summer.replace({'gender':{'女':0}},inplace = True)
summer.replace({'gender':{'男':1}},inplace = True)


```
