```python
'''
    读取2021年夏季数据并分别查看
    bmi > 24
    18 < bmi < 24
    bmi > 24
    的pmv分布
'''
df = pd.read_csv('../../dataset/summer/summer_2021_by_person.csv',encoding="gbk")
df.dropna(axis=0,how='any',inplace = True)

data = df.loc[df["bmi"] <= 18]
pmv_distribution = sns.kdeplot(data['thermal sensation'], shade=True)
pmv_distribution.axes.set_title(' thin distribution plot of 2021 summer', fontsize=10)
pmv_distribution.set_xlabel('pmv', fontsize=10)
pmv_distribution.set_ylabel('density', fontsize=10)
plt.savefig('.././result/summer/scatter_bmi/thin distribution plot of 2021 summer',dpi=200, bbox_inches='tight')
plt.show()

data = df.loc[(df["bmi"] > 18) & (df["bmi"] < 24)]

pmv_distribution = sns.kdeplot(data['thermal sensation'], shade=True)
pmv_distribution.axes.set_title(' medium distribution plot of 2021 summer', fontsize=10)
pmv_distribution.set_xlabel('pmv', fontsize=10)
pmv_distribution.set_ylabel('density', fontsize=10)
plt.savefig('.././result/summer/scatter_bmi/medium distribution plot of 2021 summer',dpi=200, bbox_inches='tight')
plt.show()

data = df.loc[df["bmi"] >= 24]
pmv_distribution = sns.kdeplot(data['thermal sensation'], shade=True)
pmv_distribution.axes.set_title(' fat distribution plot of 2021 summer', fontsize=10)
pmv_distribution.set_xlabel('pmv', fontsize=10)
pmv_distribution.set_ylabel('density', fontsize=10)
plt.savefig('.././result/summer/scatter_bmi/fat distribution plot of 2021 summer',dpi=200, bbox_inches='tight')
plt.show()

```
