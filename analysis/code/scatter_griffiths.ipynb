```python
# 读取2021年全年数据，查看每个人员的格里菲斯常数


dfs = pd.read_csv('../../dataset/summer/summer_2021_by_person.csv',encoding="gbk")
dfw = pd.read_csv('../../dataset/winter/winter_2021_by_person.csv',encoding="gbk")

dfs.dropna(axis=0,how='any',inplace = True)
dfw.dropna(axis=0,how='any',inplace = True)

# pmv_summer = dfs.loc[:,['thermal sensation']];
# pmv_winter = dfw.loc[:,['thermal sensation']];
#
# pmv_all = pd.concat([pmv_summer,pmv_winter])
#
#
# temp_summer = dfs.loc[:,['temp']]
# temp_winter = dfw.loc[:,['temp']]
#
# temp_all = pd.concat([temp_summer,temp_winter])
#
# # scatter = sns.scatterplot(x = pmv_all,y = temp_all)
# # scatter.set_xlabel('pmv')
# # scatter.set_ylabel('temp')
# statistics = linregress(pmv_all,temp_all)
# print("群体pmv与temp的线性关系为：")
# print(statistics)

for i in range(len(names)):
    data_summer = dfs.loc[dfs["no"] == i]
    data_winter = dfw.loc[dfw["no"] == i]

    data_summer = data_summer[['thermal sensation','temp']]
    data_winter = data_winter[['thermal sensation','temp']]
    data = pd.concat([data_summer,data_winter])
    '''
    LinregressResult(
        slope:斜率
        intercept:截距
        rvalue:相关性,
        pvalue=3.1209429808110146e-143,
        stderr:标准误差
        intercept_stderr:截距的标准误差
    )
    '''

    if len(data) != 0:
        temp = data['temp']
        pmv = data['thermal sensation']
        scatter = sns.scatterplot(x = pmv,y = temp)
        scatter.set_xlabel('pmv')
        scatter.set_ylabel('temp')
        statistics = linregress(pmv,temp)
        print(statistics)
        # print(str(i) + '全年pmv与temp的相关性系数为' + str(statistics.rvalue))

```
